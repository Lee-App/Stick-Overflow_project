```
import numpy as np
import time

#상관계수함수 
def cor(lst1,lst2):
  
  #데이터 길이 맞지 않을 경우 처리
  #if(len(lst1)!=len(lst2)):
  #  print('두 데이터의 길이가 다릅니다.')
  #  return
  
  #변수 데이터 초기화 및 평균값 계산
  sum1,sum2,sum3 = 0,0,0
  avg = np.mean(lst1)
  avg2 = np.mean(lst2)
  
  #수학적 공식 계산
  for i in range(len(lst1)):
    sum1 = sum1 + pow(lst1[i]-avg,2)
    sum2 = sum2 + pow(lst2[i]-avg2,2)
    sum3 = sum3 + ((lst1[i]-avg)*(lst2[i]-avg2))
  result = sum3/pow((sum1*sum2),1/2)
  
  return "%0.2f" % result

start_time = time.time()


a = df['SepalLengthCm']
b = df['SepalWidthCm']
start_time = time.time()
r = cor(a,b)
end_time = time.time()
print(end_time-start_time)
start_time = time.time()
df2 = pd.DataFrame({'A':a,'B':b})
result = df2.corr()
end_time = time.time()
print(end_time-start_time)
print(r)
print(result)
```
